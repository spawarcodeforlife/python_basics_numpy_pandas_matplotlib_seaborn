```python
x = [0,1,2,3,4]

y = [0,2,4,6,8]

plt.plot(x,y, 'r.--', label = '2x')


x2  = np.arange(0,4.5,0.5)
plt.plot(x2, x2**2, 'blue', label ='x^2')

plt.title('My first graph!', fontdict ={'fontname':'comic sans ms', 'fontsize': 21})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

```python
x = [0,1,2,3,4]

y = [0,2,4,6,8]

plt.plot(x,y, 'r.--', label = '2x')


x2  = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'blue', label ='x^2')
plt.plot(x2[5:], x2[5:]**2, 'b--')

plt.title('My first graph!', fontdict ={'fontname':'comic sans ms', 'fontsize': 21})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

```python
x = [0,1,2,3,4]

y = [0,2,4,6,8]

plt.figure(figsize=(5,3), dpi = 300)

plt.plot(x,y, 'r.--', label = '2x')


x2  = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'blue', label ='x^2')
plt.plot(x2[5:], x2[5:]**2, 'b--')

plt.title('My first graph!', fontdict ={'fontname':'comic sans ms', 'fontsize': 21})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.show()
```

```python
x = [0,1,2,3,4]

y = [0,2,4,6,8]

plt.figure(figsize=(5,3), dpi = 300)

plt.plot(x,y, 'r.--', label = '2x')


x2  = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'blue', label ='x^2')
plt.plot(x2[5:], x2[5:]**2, 'b--')

plt.title('My first graph!', fontdict ={'fontname':'comic sans ms', 'fontsize': 21})
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()

plt.savefig('mygraph.png', dpi = 300)

plt.show()
```

```python
labels = ['A','B', 'C']
values = [1,4,2]

plt.bar(labels, values)

plt.show()
```

```python
labels = ['A','B', 'C']
values = [1,4,2]

plt.bar(labels, values)

plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ['A','B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch("*")

plt.figure(figsize=(6,4))

plt.show()
```

```python
labels = ['A','B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))


plt.figure(figsize=(6,4))

plt.show()
```
